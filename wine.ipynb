```python
df.groupby("Color")["quality"].mean()
```

```python
df.rename(columns= lambda x:x.replace(" ", "_"),inplace=True)
df.head()
```

```python
bing = [2.72, 3.11, 3.21, 3.32, 4.01]
label = ["High", "Mod_High", "Medium", "Low"]
```

```python
df["level_of_acidity"] = pd.cut(df['pH'], bing, labels=label)
df.head()
```

```python
df.groupby('level_of_acidity')['quality'].mean()
```
